```python
import pandas as pd

data = pd.read_csv('Dataset Twitter Fix - Indonesian Sentiment Twitter Dataset Labeled (1).csv')
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from gensim.utils import simple_preprocess

stop_words = StopWordRemoverFactory().get_stop_words()

def preprocess(text):
    return [word for word in simple_preprocess(text) if word not in stop_words]

data['preprocessed'] = data['Tweet'].apply(preprocess)
```

```python
data
```

```python
from gensim.models import Word2Vec

sentences = data['preprocessed'].tolist()

word2vec_model = Word2Vec(sentences, vector_size=300, window=5, min_count=1, workers=4)

word2vec_model.save("word2vec_custom_model.model")
```
